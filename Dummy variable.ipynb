```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r"C:\Users\Usman\Downloads\page2\carprices.csv")
```

```python
df
```

```python
a=pd.get_dummies(df[["Car Model"]])
a
```

```python
merge=pd.concat([df,a],axis=1)
merge
```

```python
finaldf=merge.drop(["Car Model_Mercedez Benz C class","Car Model"],axis=1)
finaldf
```

```python
reg=LinearRegression()
```

```python
X=finaldf.drop(["Sell Price($)"],axis=1)
X
```

```python
y=finaldf[["Sell Price($)"]]
y
```

```python
reg.fit(X,y)
```

```python
reg.predict([[45000,4,0,0]])
```

```python
reg.predict([[86000,7,0,1]])
```

```python
reg.score(X,y)
```
